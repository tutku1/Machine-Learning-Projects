```python

df[['Gender', 'Height', 'Weight']] = df['Gender,"Height","Weight"'].str.split(',', expand=True)

df['Height'] = pd.to_numeric(df['Height'])
df['Weight'] = pd.to_numeric(df['Weight'])
```

```python
df
```

```python
df.drop(columns=['Gender,"Height","Weight"'], inplace=True)
df
```

```python
df.Height.describe()
```

```python
import numpy as np

height_median = np.median(df.Height)
height_std = np.std(df.Height)
height_mean = np.mean(df.Height)
weight_median = np.median(df.Weight)
weight_std = np.std(df.Weight)
weight_mean = np.mean(df.Weight)
print(height_median, height_std, height_mean, weight_median, weight_std, weight_mean)
```

```python
upper_limit = height_mean + 3*height_std
print(upper_limit)
lower_limit = height_mean - 3*height_std
print(lower_limit)
```

```python
df[(df.Height>upper_limit) | (df.Height<lower_limit)]
```

```python
df_no_outlier_std_dev = df[(df.Height<upper_limit) & (df.Height>lower_limit)]
df_no_outlier_std_dev.head()
```
